#### 1. Convert the input text to a list of sentences. Then, compute the number of sentences in the given Text.

```python
import nltk
import math
#nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
import nltk.data
from nltk import tokenize
with open('Article.txt', 'r') as file:
    data = file.read()
    data=data.replace('Â','')
    data=data.replace('\xa0',' ')
    data=data.replace('â€”','')
    data=data.replace('.  ','. ')
data=data.lower()
text_lines=tokenize.sent_tokenize(data)
text_lines
```

#### 2. Convert the input text to a list of sentences. Then, compute the number of sentences in the given Text.

```python
word_freq={}
for line in text_lines:
    dict_1={}
    nline=line.split()
    for word in nline:
        word=word.lower().strip().replace('"','').replace(',','')
        dict_1[word]=nline.count(word)/len(nline)
    word_freq[line]=dict_1
word_freq
```

#### 3. Calculate Term frequency for each word in a sentence: 
#### TF(word)=(Number of times term “word” appears in a sentence) / (Total number of terms in the sentence)
#### 4. Create a matrix termFrequency: The termFrequency matrix should be a dictionary where each key is a sentence and the value is also a dictionary of word frequency.

```python
term_freq_dict={}
for line in text_lines:
    dic3={}
    nline=line.split()
    for word in nline:
        word=word.lower().strip().replace('"','').replace(',','')
        dic3[word]=nline.count(word)/len(nline)
    term_freq_dict[line]=dic3
term_freq_dict
```

#### 5. For each word compute how many sentences contain that word.


```python
data_sorted=sorted(set(data.lower().strip().split()))
data_lower=[x.lower() for x in text_lines]
word_per_sentence={}
lst=[]
for word in data_sorted:
    for j in range(len(data_lower)):
        if word in sorted(set(data_lower[j].split())):
            word = word.strip().replace('"','').replace(',','')
            lst.append(word)
            word_per_sentence[word]=lst.count(word)
word_per_sentence
```

#### 6. Calculate IDF for each word in a sentence.
#### IDF(word) = log e(Total number of sentences/ number of sentences with term word in it)


```python
IDF_dict={}
for line in text_lines:
    dic6={}
    nline=line.split()
    for word in nline:
        word=word.lower().strip().replace('"','').replace(',','')
        dic6[word]=round(math.log10(len(text_lines)/word_per_sentence[word]),2)
    IDF_dict[line]=dic6
IDF_dict
```

#### 7. Compute the TF-IDF for each word in each sentence.

```python
TF_IDF_dict={}
for line in text_lines:
    dic8={}
    nline=line.split()
    for word in nline:
        word=word.lower().strip().replace('"','').replace(',','')
        dic8[word]=round((word_freq[line][word]*IDF_dict[line][word]),2)
    TF_IDF_dict[line]=dic8
TF_IDF_dict
```

#### 8. Use the TF-IDF computed in (7) and give a weight for each sentence.


```python
weights={}
for line in text_lines:
    nline=line.split()
    weight=0
    count=0
    for word in nline:
        word=word.lower().strip().replace('"','').replace(',','')
        weight=weight+TF_IDF_dict[line][word]
        count=count+1
    weights[line]=round(weight/count,2)
weights
```

#### 9. Threshold: compute the average sentence weight


```python
summ=0
for x,y in weights.items():
    summ=summ+y
threshold=summ/79
threshold
```

#### 10. Generate the summary : select a sentence for summarization if the weight of the sentence exceeds the threshold.

```python
summary=""
counts=0
for x,y in weights.items():
    if y>threshold:
        summary=summary+" "+x
        counts=counts+1
print(summary.strip())
```
